```python
import warnings # 忽略警告
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
warnings.filterwarnings('ignore') # 忽略警告
```

```python
data = pd.read_csv(
'https://labfile.oss.aliyuncs.com/courses/1283/adult.data.csv')

data.head()
```

## 数据集中有多少男性和女性？

```python
data['sex'].value_counts()
```

## 数据集中女性的平均年龄是多少？

```python
df_female = data[data['sex'] == 'Female']
female_mean_age = df_female['age'].mean()
print(female_mean_age)
```

## 数据集中德国公民的比例是多少？

```python
float((data['native-country'] == 'Germany').sum()) / data.shape[0]
```

## 年收入超过 50K 和低于 50K 人群年龄的平均值和标准差是多少？

```python
gt_50 = data[data['salary'] == '>50K']['age']
lt_50 = data[data['salary'] == '<=50K']['age']
gt_50_mean = gt_50.mean()
lt_50_mean = lt_50.mean()
gt_50_std = gt_50.std()
lt_50_std = lt_50.std()
print('rich {},{}'.format(round(gt_50_mean),round(gt_50_std,1)))
print('poor {},{}'.format(round(lt_50_mean),round(lt_50_std,1)))
```

## 年收入超过 50K 的人群是否都接受过高中以上教育？

```python
data[data['salary'] == '>50K']['education'].unique()
```

## 使用 groupby 和 describe 统计不同种族和性别人群的年龄分布数据。

```python
for (race,sex), sub_df in data.groupby(['race','sex']):
    print("Race: {}, sex: {}".format(race,sex))
    print(sub_df['age'].describe())
```

## 统计男性高收入人群中已婚和未婚（包含离婚和分居）人群各自所占数量。

```python
data[(data['salary'] == '>50K') & (data['sex'] == 'Male')]['marital-status'].value_counts()
```

## 统计数据集中最长周工作小时数及对应的人数，并计算该群体中收入超过 50K 的比例。

```python
max_load = data['hours-per-week'].max()
print("Max time {} hours./week.".format(max_load))
num_workabolics = data[data['hours-per-week'] == max_load].shape[0]
print('Total number of such hard workers {}'.format(num_workabolics))
rich_share = float(data[(data['hours-per-week'] == max_load) & (data['salary'] == '>50K')].shape[0]) / num_workabolics
print("Percentage of rich among them {}%".format(int(rich_share * 100)))
```

## 计算各国超过和低于 50K 人群各自的平均周工作时长。

```python
countries = data['native-country'].unique()
for country in countries:
    country_hours_per_week = data[data['native-country'] == country].groupby(['salary'])['hours-per-week'].mean()
    print('Country : {} \n {}'.format(country,round(country_hours_per_week,2)))
```

```python
for (country,salary), sub_df in data.groupby(['native-country','salary'])['hours-per-week']:
    print(country,salary,round(sub_df.mean(),2))
```
